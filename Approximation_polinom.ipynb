```python
import scipy
import math
import matplotlib.pyplot as plt
# f(x) = sin(x / 5) * exp(x / 10) + 5 * exp(-x / 2)
# y = w0 + w1*x
```

```python
def fx(x):
    y=np.sin(x/5)*np.exp(x/10)+5*np.exp(-x/2)
    return y
x=np.arange(1,16)
y=np.sin(x/5)*np.exp(x/10)+5*np.exp(-x/2)

print('Многочлен первой степени:')
print('Значение функции при x=1: ',fx(1))
print('Значение функции при x=15: ',fx(15))

# w0 + 1 * w1 = 3.252
# w0 + 15 * w1 = 0.635

A = np.array([[1, 1], [1, 15]])
B = np.array([fx(1),fx(15)])
sol1=scipy.linalg.solve(A,B)
print('y='+str(sol1[0])+str(sol1[1])+'x')

fig = plt.subplots()
x1=np.linspace(0,16,4)
y1= sol1[0]+sol1[1]*x1
plt.plot(x1, y1)
plt.plot(x,y)
plt.show()
```

```python
print('Многочлен второй степени:')
print('Значение функции при x=1: ',fx(1))
print('Значение функции при x=8: ',fx(8))
print('Значение функции при x=15: ',fx(15))

# w0 + 1 * w1 + 1 * w2 = 3.252
# w0 + 8 * w1 + 64 * w2 = 2.317
# w0 + 15 * w1 + 225 * w2 = 0.635

A = np.array([[1,1,1],[1,8,64], [1,15,225]])
B = np.array([fx(1),fx(8),fx(15)])
sol2=scipy.linalg.solve(A,B)
print('y='+str(sol2[0])+str(sol2[1])+'x'+str(sol2[1])+'x^2')

fig = plt.subplots()
x1=np.linspace(0,16,4)
y1= sol2[0]+sol2[1]*x1+sol2[2]*x1**2
plt.plot(x1, y1)
plt.plot(x,y)
plt.show()
```

```python
print('Многочлен третьей степени:')
print('Значение функции при x=1: ',fx(1))
print('Значение функции при x=4: ',fx(4))
print('Значение функции при x=10: ',fx(10))
print('Значение функции при x=15: ',fx(15))

# w0 + 1 * w1 + 1 * w2 + 1 * w3  = 3.252
# w0 + 4 * w1 + 16 * w2 + 64 * w3  = 1.747
# w0 + 10 * w1 + 100 * w2 + 1000 * w3  = 2.505
# w0 + 15 * w1 + 225 * w2 + 3375 * w3  = 0.635

A = np.array([[1,1,1,1],[1,4,16,64],[1,10,100,1000], [1,15,225,3375]])
B = np.array([fx(1),fx(4),fx(10),fx(15)])
sol3=scipy.linalg.solve(A,B)
print('y='+str(sol3[0])+str(sol3[1])+'x'+str(sol3[2])+'x^2'+str(sol3[3])+'x^3')

fig = plt.subplots()
x1=np.linspace(0,16,4)
y1= sol3[0]+sol3[1]*x1+sol3[2]*x1**2+sol3[3]*x1**3
plt.plot(x1, y1)
plt.plot(x,y)
plt.show()
```
